```python
import serpapi


import requests


def google_custom_search(query, api_key, cx, page_num):
  url = f"https://www.googleapis.com/customsearch/v1"
  as_nlo = page_num * 10
  as_nhi = page_num * 10 + 9
  params = {
    'key': api_key,
    'cx': cx,
    'q': query,
    'start': page_num * 10 + 1
    # 'as_nlo': as_nlo,
    # 'as_nhi': as_nhi
    
  }
  try:
      # pass
    response = requests.get(url, params=params)
    data = response.json()
    return data
  except Exception as e:
    print(f"An error occurred: {str(e)}")
    return None
```

```python


#help from https://stackoverflow.com/questions/55866199/get-more-than-10-results-google-custom-search-api
def search(query, api_key, cx, page_num):
  # Query to search
  search_results = google_custom_search(query, api_key, cx, page_num)
  if search_results:
    print(search_results)
    # for item in search_results['items']:
      # print(item['title'])
      # print(item['link'])
      # print(item['snippet'])
      # print()
    return search_results
  else:
    print("No search results found.")
```

```python
import os.path
import json


def search_n_pages(api_key, cx, pages):
  query = input("Enter your search query: ")
  for i in range(pages):
    print(i)
    search_results = search(query, api_key, cx, i)
    if not os.path.exists(os.path.join(os.getcwd(), query)):
      os.makedirs(os.path.join(os.getcwd(), query))
    path = os.path.join(query,  query + str(i) + '.json')
    with open(path, 'w') as fp:
      json.dump(search_results, fp)
    
```
